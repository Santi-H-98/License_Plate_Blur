```python
import os
from ultralytics import YOLO 
import cv2
```

```python
model_best_path = "C:/Users/santi/Desktop/PROGRAMACION/COMPUTER VISION/Projects/License Plate Detector/yolo-license-plate-detection/runs/detect/train2/weights/last.pt"
model = YOLO(model_best_path)
```

```python
demo_test_path = "C:/Users/santi/Desktop/PROGRAMACION/COMPUTER VISION/Projects/License Plate Detector/yolo-license-plate-detection/data"

for filename in os.listdir(demo_test_path):
    if filename.endswith('.jpg'):

        image_path = os.path.join(demo_test_path,filename)
        results = model.predict(image_path ,visualize=True ,conf=0.25,augment=True)

        for result in results:
            boxes = result.boxes
            # Establezco directorio de saluda
            output_directory = "C:/Users/santi/Desktop/PROGRAMACION/COMPUTER VISION/Projects/License Plate Detector/yolo-license-plate-detection/data/predictions"
            output_filename = filename + '_detections.jpg'
            output_path = os.path.join(output_directory, output_filename)
            result.save(output_path,output_filename)
```
